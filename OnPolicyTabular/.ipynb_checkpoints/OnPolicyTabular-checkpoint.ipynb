```python
'''
Comparitive Mean-variance Tradeoff Plots for VPAC
'''

mpl.style.use('classic')
plt.rcParams.update({'font.size':15})
fig, (ax0, ax1) = plt.subplots(ncols=2,sharex=True, figsize=(10, 4))
plt.tight_layout()

major_dir = "../../Results/Results_FR" # Add directory name where results are stored
path = os.path.join(major_dir, "FourRoom") # Path for reading data. Check code for changing path dir.
save_folder_plot = os.path.join(major_dir,"FourRoomPlots")
if not os.path.exists(save_folder_plot):
        os.makedirs(save_folder_plot)

psi_list = ["0.0","0.01", "0.03", "0.05"]
line_labels = []
dirs = os.listdir(path)
file_names_to_read = []
for psi in psi_list:
    for dir_name in dirs:
        if (dir_name.startswith("R100_")and "seed10" in dir_name and "lambda1.0" in dir_name 
            and "temp1.0" in dir_name):
            psi_val = dir_name.split("_")[2].split("Psi")[1]
            if psi_val != psi:
                continue
            files = os.listdir(os.path.join(path, dir_name))
            if "History.npy" in files:
                file_names_to_read.append(dir_name)
                line_labels.append(r'$\psi$='+psi_val)
            
  

x_axis = list(np.arange(0,1000,50))
x_axis.append(1000-1)
x_axis = np.array(x_axis)
line_object =[]
for ind,f in enumerate(file_names_to_read):
    psi = f.split("_")[2].split("Psi")[1]
    print("ind: ", ind, "--------")
    print(psi)
    var_mean = np.load(os.path.join(path, f, 'RewardDistributionVarMean.npy'))
    var_std = np.load(os.path.join(path, f, 'RewardDistributionVarStd.npy'))
    mean_mean = np.load(os.path.join(path, f, 'RewardDistributionMeanMean.npy'))
    mean_std = np.load(os.path.join(path, f, 'RewardDistributionMeanStd.npy'))
    
    new_mean = []
    window = 50
    for eps in np.arange(0,1000,window):
        new_mean.append(np.mean(mean_mean[eps: eps+window]))
    linestyle = "-"
    print(linestyle + f)
    
    print("mean: ", np.mean(mean_mean[:-20]), " var: ", var_mean[-1])
    l = ax0.plot(np.arange(0,1000,window), new_mean, label = f,linewidth = 2, linestyle = linestyle)
    ax1.plot(x_axis, var_mean, label = f,linewidth = 2, linestyle = linestyle)
    line_object.append(l)
    
fig.legend(line_object,labels=line_labels,   # The labels for each line
           loc="best",   # Position of legend
           borderaxespad=0.1,
          prop={'size': 13})

ax0.set_ylabel("Return",fontsize=18)
ax1.set_ylabel("Variance in Return",fontsize=18)
ax0.set_xlabel("Episode",fontsize=18)
ax1.set_xlabel("Episode",fontsize=18)
ax0.grid(True)
ax1.grid(True)

fig.savefig((os.path.join(save_folder_plot,"ComparisonPsi"+".png")), dpi=200, facecolor='w', edgecolor='w',
           orientation='portrait', papertype=None, format=None,
           transparent=True, bbox_inches='tight', pad_inches=0,
           frameon=None)        

```

```python
# Plots for fourrooms tabular setting
major_dir = "../../Results/Results_FR"    

names = ["VAPC_Psi0.0", "VPAC_Psi0.015"] # add dir from where to read the data for VPAC psi=0, and psi=0.015

env = gym.make('Fourrooms-v0')
num_states = env.observation_space.n
features = Tabular(num_states)
nactions = env.action_space.n
weight_policy_total_list=[]
for index in range(len(names)):
    weight_policy_total_list.append(np.load(os.path.join(major_dir,names[index],"Weights_Policy.npy")))
 
```

```python
def getEnvLayout(layout, ann_format_char = True, plot_type_policy = "True", wall_color = 0):
    num_elem = 13
    line_count = 0
    mat_layout = -2*np.ones((num_elem, num_elem))
    mapping_state_row_col = defaultdict(list)
    ann_layout_char = np.chararray((num_elem, num_elem), unicode=True)
    ann_layout_num = np.zeros((num_elem, num_elem))            
    state_count = 0
    for line in layout.splitlines():
        for i in range(num_elem):
            if line[i]!="w":
                mat_layout[line_count, i] = 5
                mapping_state_row_col[state_count].extend([line_count, i])
                ann_layout_num[line_count, i] = state_count
                state_count +=1            
            if line[i] == "f":                
                mat_layout[line_count, i] = 10
                ann_layout_char[line_count, i]='F'
            elif line[i]=="w":
                mat_layout[line_count, i] = wall_color
        line_count +=1
    if ann_format_char==True:
        ann_layout = ann_layout_char
    else:
        ann_layout = ann_layout_num
    return mat_layout, ann_layout, mapping_state_row_col
```

```python
def GetFrozenStates():
    num_elem = 13
    line_count = 0
    element_count = 0
    frozen_states =[]
    state_num=0
    for line in layout.splitlines():
        for i in range(num_elem):        
            if line[i] == "f":
                frozen_states.append(state_num)
            if line[i]!="w":
                state_num +=1
    return frozen_states
```

```python
# Visulaization of how fourrooms env looks

plt.figsize=(1,1)
mat_layout,ann_layout, mapping_state_row_col = getEnvLayout(layout, False)
sn.set(font_scale=0.6)
ax = sn.heatmap(mat_layout, annot = ann_layout, fmt = '', linewidth =0.5, linecolor = "white")
plt.show()
plt.close()
```

## Sampled policy trajectories 

```python
init_state_done =[]
count_graph_plot = 0
run_to_show = run_val # run number of experiment
episode =-1
folder = os.path.join(r_folder,"traj")
if not os.path.exists(folder):
        os.makedirs(folder)

folder = os.path.join(folder,names[safe])
if not os.path.exists(folder):
    os.makedirs(folder)
        
num_elem = 13
possible_goal_states = [62] # goal location
max_time_step = 50
# 0:U
# 1:D
# 2:L
# 3:R
# 4:UR
# 5:DR
# 6:DL
# 7:UL
action_mapping = [u'↑',u'↓', u'←',u'→',u'↗',u'↘',u'↙',u'↖']

start_states =[]
num_traj =103
count_traj = 0
while(count_traj<num_traj):    
    curr_time = 0
    init_state = env.reset()
    if init_state in init_state_done:
        continue
    else:
        count_traj +=1
        ann_action = np.chararray((num_elem, num_elem), unicode=True)
        init_state_done.append(init_state)
        

    print("init state: ", init_state)
    curr_state = init_state
    cmap = sn.dark_palette("seagreen",n_colors=4, as_cmap=True)

    goal_state = 62
    mat_layout, ann_layout, mapping_state_row_col = getEnvLayout(layout, True, False, -0.1)
    while(curr_state!= goal_state and max_time_step > curr_time): 
        action = np.argmax(weight_policy_total[run_to_show,episode,curr_state, :])
        next_state,r,d,_ = env.step(action)
        row, col = mapping_state_row_col[curr_state]
        row_2, col_2 = mapping_state_row_col[next_state]
        mat_layout[row, col] = 15
        if curr_state == init_state:
            ann_action[row,col] = 'S'
            mat_layout[row,col] = 20
        else:
            ann_action[row, col] =action_mapping[action]
        curr_state = next_state
        curr_time +=1

    if curr_state == possible_goal_states[0]:
        r,c = mapping_state_row_col[curr_state]
        mat_layout[r,c] = 20
        ann_action[r,c] = 'G'
        
    plt.figure(figsize = (5,5))
    sn.set(font_scale=0.6)
    ax = sn.heatmap(mat_layout, annot = ann_action, fmt = '', 
                    cbar=False,annot_kws={"size": 20},square=True,
                    mask=mat_layout<0,
                    linewidth =0.25, linecolor = "grey")
    
    ax.set_facecolor('black')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    init_state_done.append(init_state)
    count_graph_plot +=1
    
    plt.savefig((os.path.join(folder, "Policy_init"+str(init_state)+"_run"+str(run_to_show)+".png")), dpi=300, facecolor='k', 
    edgecolor='k', orientation='portrait', papertype=None, format=None,
    transparent=True, bbox_inches='tight', pad_inches=0, frameon=None)
    
    plt.show()
    plt.close()
```

## Variance in return for each state in fourrooms as start state

```python
'''
Variance Plot for initial state distribution
'''

run_to_show = run_val
dir_name = os.path.join(r_folder,"Variance")
if not os.path.exists(dir_name):
        os.makedirs(dir_name)
      
folder = os.path.join(dir_name,names[safe])
if not os.path.exists(folder):
        os.makedirs(folder)
        
n_start_states = 104
episode =-1
freq_each_state_visit = 60

reward_distribution_states_variance = np.zeros(n_start_states)
freq_visit_states = np.zeros(n_start_states)
reward_distribution = np.zeros((n_start_states,freq_each_state_visit))
max_time_step = 200
frozen_states = GetFrozenStates()
n_trajectories = 40000
c_trajectories = 0

while(np.sum(freq_visit_states > freq_each_state_visit)!= n_start_states-1 and c_trajectories< n_trajectories):
    c_trajectories +=1
    start = env.reset()
    if freq_visit_states[start] >= freq_each_state_visit:
        continue
    freq_visit_states[start] +=1    
    curr_state = start
    curr_time = 0
    d= False
    gamma=0.99
    current_dicounting_factor=1.
    discounted_reward = 0
    while(curr_state!= 62 and max_time_step > curr_time and d!=True):
        action = np.argmax(weight_policy_total[run_to_show,episode,curr_state, :])
        next_state,r,d,_ = env.step(action)
        if curr_state in frozen_states:
            r = np.random.normal(0,8.0)
        discounted_reward += current_dicounting_factor*r
        curr_state = next_state
        current_dicounting_factor*=gamma
    reward_distribution[start,int(freq_visit_states[start])-1] = discounted_reward
    
new_reward_var = np.var(reward_distribution,axis=1)
new_reward_mean = np.mean(reward_distribution,axis=1)
```

```python
new_reward_var[new_reward_var>700]=700
mat_layout, ann_layout, mapping_state_row_col = getEnvLayout(layout, True, wall_color= -0.01)
for i in range(n_start_states):
    row, col = mapping_state_row_col[i]
    mat_layout[row, col] = np.round(new_reward_var[i],2)
r,c = mapping_state_row_col[62]
ann_layout[r,c] = 'G'

plt.figure(figsize = (5,5))
sn.set(font_scale=1)
ax = sn.heatmap(mat_layout, annot=ann_layout,fmt = '', cmap="YlGnBu",
                vmin = 0, cbar = True, annot_kws={"size":14}, cbar_kws={"shrink": .7},
                square=True, linewidth =0.25, linecolor = "grey", mask=mat_layout<0)
ax.set_facecolor('black')
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
save_name = "Var.png"

plt.savefig((os.path.join(folder,save_name)), dpi=100, facecolor='w', edgecolor='k',
           orientation='portrait', papertype=None, format=None,
           transparent=True, bbox_inches='tight', pad_inches=0,
           frameon=None)
plt.show()
plt.close()
```
